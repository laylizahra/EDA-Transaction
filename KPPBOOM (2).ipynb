```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Basic Info ---
print("ðŸ”Ž Info Dataset")
print(df.info())
print("\nðŸ“Œ Statistik Deskriptif")
print(df.describe(include='all'))

# --- Missing Values ---
print("\nâ“ Missing Values")
print(df.isnull().sum())

# --- Distribusi Numerical ---
num_cols = df.select_dtypes(include=['int64','float64']).columns
print("\nðŸ“Š Kolom Numerik:", num_cols.tolist())

for col in num_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Distribusi {col}')
    plt.xlabel(col)
    plt.ylabel("Frekuensi")
    plt.show()

# --- Distribusi Categorical ---
cat_cols = df.select_dtypes(include=['object']).columns
print("\nðŸ”  Kolom Kategorikal:", cat_cols.tolist())

for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f'Frekuensi {col}')
    plt.show()

# --- Korelasi Numerical ---
if len(num_cols) > 1:
    plt.figure(figsize=(10,6))
    sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Heatmap Korelasi Numerik")
    plt.show()

# --- Boxplot Amount ---
amount_cols = [col for col in df.columns if 'amount' in col.lower()]
if amount_cols:
    amount_col = amount_cols[0]
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[amount_col])
    plt.title(f'Boxplot {amount_col}')
    plt.show()
```
